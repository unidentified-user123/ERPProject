```python
arr=[1,4,6,7,9,-1]
if not arr:
  print("empty")
elif len(arr)<2:
  print("array must be at least two elements")
else:
  largest_num=float('-inf')
  second_largest=float('-inf')
  for number in arr:
    if number>largest_num:
      second_largest=largest_num
      largest_num=number
    elif number > second_largest and number!=largest_num:
      second_largest=number

  if second_largest==float('-inf'):
    print("no distinct second largest number found")
  else:
    print(second_largest)
```

```python
number=[2,3,4,5,6,78]

# Initialize largest and second to very small values
largest = float('-inf')
second = float('-inf')

# Iterate through each element in the list 'number'
for num_val in number: # Renamed loop variable to avoid conflict with list name
    if num_val > largest:
        second = largest
        largest = num_val
    elif num_val > second and num_val != largest:
        second = num_val

# After checking all numbers, print the second largest
# Add checks for edge cases similar to the previous example
if second == float('-inf'):
    print("No distinct second largest number found (e.g., array too small or all elements are same)")
else:
    print(second)
```
